Gráficos de análisis univariados
Gráficos que se pueden utilizar para el análisis de variables numéricas

```python
# Seleccionar las columnas 
columns_of_interest = ['budget', 'revenue']
df_selected_columns = df[columns_of_interest]

# Crear histograma para 'budget'
plt.figure(figsize=(10, 6))
plt.hist(df_selected_columns['budget'], bins=20, color='blue', alpha=0.7)
plt.xlabel('Presupuesto')
plt.ylabel('Frequency')
plt.title('Distribucion de Presupuestos')
plt.grid(True)
plt.show()

# Crear histograma para 'revenue'
plt.figure(figsize=(10, 6))
plt.hist(df_selected_columns['revenue'], bins=20, color='green', alpha=0.7)
plt.xlabel('Ganancias')
plt.ylabel('Frequency')
plt.title('Distribucion de Ganancias')
plt.grid(True)
plt.show()
```

Diagramas de dispersión
Un diagrama de dispersión representa la asociación entre dos variables, si las variables tienden a aumentar o disminuir juntas, se dice que la asociación es positiva. Si una variable aumenta y otra disminuye, se dice que tiene una relación negativa. Si no hay un patrón, la relación es cero.

```python
sns.scatterplot(x=df['budget'], y=df['revenue'])
```

Agrupación de datos

```python
df.groupby(["belongs_to_collection"])["revenue"].mean()
```

```python
df.groupby(['belongs_to_collection','original_language'])['revenue'].mean().unstack(1).plot.barh()
```

```python
df.groupby(["belongs_to_collection"])["budget"].mean()
```

```python
df.groupby(["belongs_to_collection","original_language"])["budget"].mean().unstack(1).plot.barh()
```

Correlación

```python
plt.figure(figsize=(15,7))
# df.corr() creates a correlation matrix 
corr_matrix = df.corr()
# helps in creating a 0's matrix of df.corr() shape
mask = np.zeros_like(corr_matrix)
# Returns copy of array with lower part of the triangle
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    sns.heatmap(corr_matrix, mask=mask, square=True)
```

```python
plt.figure(figsize=(15,7))
# df_raw.corr() creates a correlation matrix 
corr_matrix = df_raw.corr()
# helps in creating a 0's matrix of df.corr() shape
mask = np.zeros_like(corr_matrix)
# Returns copy of array with lower part of the triangle
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    sns.heatmap(corr_matrix, mask=mask, square=True)
```
