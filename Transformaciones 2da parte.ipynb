Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

```python

# Leer el Dataset 
file_path = "movies_dataset.csv"
df = pd.read_csv(file_path)
# Calcula el ROI y asigna 0 cuando revenue o budget es 0 o no está presente


# Convierte las columnas "revenue" y "budget" al tipo numérico adecuado y reemplaza NaN con 0
df["revenue"] = pd.to_numeric(df["revenue"], errors="coerce").fillna(0)
df["budget"] = pd.to_numeric(df["budget"], errors="coerce").fillna(0)

# Calcula el ROI y asigna 0 cuando revenue o budget es 0
def calculate_roi(row):
    if row["budget"] != 0:
        return row["revenue"] / row["budget"]
    else:
        return 0

# Crea la nueva columna "return" aplicando la función calculate_roi a cada fila
df["return"] = df.apply(calculate_roi, axis=1)

# redondear el ROI a dos decimales
df["return"] = df["return"].apply(lambda x: round(x, 2))
df.head(1)
```
